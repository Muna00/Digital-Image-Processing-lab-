```python
img_number=1
for file in glob.glob(path):
    a=cv2.imread(file)
    cv2.imshow('Original Image',a)
    c=cv2.cvtColor(a,cv2.COLOR_BGR2RGB)
    cv2.imwrite('Images/COLOR_Image'+ str(img_number)+'.jpg',c)
    img_number +=1
    cv2.imshow("Color Image",c)
    cv2.waitKey(10000)
    cv2.destroyAllWindows()

 
```
