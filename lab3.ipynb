```python
import os
import cv2
from matplotlib import pyplot as plt
from PIL import Image
```

```python
img= Image.open("Images/tiger.jpg")
```

```python
img.show()
```

```python
print(img.mode)
print(img.size)
```

```python
#rotating
rotate_img=img.rotate(40)
rotate_img.show()
```

```python
# flipping
flip_img=img.transpose(method=Image.FLIP_TOP_BOTTOM)
flip_img.show()
flip_img.save(r"FLIPPED.jpg")

flip_img1=img.transpose(method=Image.FLIP_LEFT_RIGHT)
flip_img1.show()
```

```python
# resizimng
 
size=(120,20)  
re_img=img.resize((120,20),resample=Image.BILINEAR)
re_img.show()
```

```python
#crop
width,height=img.size
left=0.05
top=height/4
right=1
bottom=5*width/2
c_img=img.crop((left,top,right,bottom))
c_img.show()
```

```python
#reading multiple files

root="NEWIM"
files=os.listdir(root)
len(files)

fig,axs=plt.subplots(nrows=1,ncols=3,figsize=(20,20))
axs=axs.flatten()
for i in range(3):
    filepath=os.path.join(root,files[i])
    imga=Image.open(filepath)
    axs[i].imshow(imga)
    axs[i].axis("off")
    axs[i].set_title(files[i])
    
plt.show()
```

```python
#converting cv2 to PIL

cv_img=cv2.imread("Images/tiger.jpg")
cv2.imshow("CV2",cv_img)
cv2.waitKey(0)
cv2.destroyAllWindows()

clr_cvt=cv2.cvtColor(cv_img,cv2.COLOR_BGR2RGB)


pil_img=Image.fromarray(clr_cvt)
pil_img.show()

```

```python
#splitting color

cv_img=cv2.imread("Images/tiger.jpg")
B,G,R= cv2.split(cv_img)

cv2.imshow("OG",cv_img)
cv2.waitKey(0)

cv2.imshow("Blue",B)
cv2.waitKey(0)

cv2.imshow("Green",G)
cv2.waitKey(0)

cv2.imshow("Red",R)
cv2.waitKey(0)


```

```python
#CMY (must be on RGB)

C=1-R
M=1-G
Y=1-B

cv2.imshow("OG",cv_img)
cv2.waitKey(0)

cv2.imshow("C",C)
cv2.waitKey(0)

cv2.imshow("M",M)
cv2.waitKey(0)

cv2.imshow("Y",Y)
cv2.waitKey(0)

cv2.destroyAllWindows()
```
