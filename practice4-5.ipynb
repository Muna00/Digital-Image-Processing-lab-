```python
import os
import cv2,glob
from matplotlib import pyplot as plt
```

```python
my_list=[]
for file in glob.glob(path):
    print(file)
    a=cv2.imread(file)
    my_list.append(a)
    
plt.imshow(my_list[2])
    
```

```python
img_no=1
for file in glob.glob(path):
    c=cv2.imread(file)
    b=cv2.cvtColor(c,cv2.COLOR_BGR2RGB)
    cv2.imwrite("NEWIM/colorImage"+str(img_no)+".jpg",b)
    img_no+=1
    cv2.imshow("color",b)
    cv2.waitKey(0)
cv2.destroyAllWindows()    
```

```python
path="Images/NEWIM/"

img_no=1
for image in os.listdir(path):
    file_path = os.path.join(path, image)
    x=cv2.imread(file_path)
    y=cv2.cvtColor(x,cv2.COLOR_BGR2RGB)
    cv2.imwrite(os.path.join,"Images/NEWIM/colorImage"+str(img_no)+".jpg",y)
    img_no+=1
    cv2.imshow("color",y)
    cv2.waitKey(0)
cv2.destroyAllWindows()    
```

```python
from skimage import io,color
from skimage.transform import rescale,resize,downscale_local_mean
```

```python
img=io.imread("Images/tiger.jpg",1)
re=rescale(img,1.0/4.0,anti_aliasing=True)
plt.imshow(re)
```

```python
resi=resize(img,(220,100),anti_aliasing=False)
plt.imshow(re)
```

```python
down=downscale_local_mean(img,(4,3))
plt.imshow(down)
```

```python
from skimage.filters import gaussian,sobel
```

```python
gaus=gaussian(img,sigma=1,mode="constant",cval=0.0)
plt.imshow(gaus)
```

```python
sob=sobel(img)
plt.imshow(sob,cmap="gray")
```

```python
#cv2 resize

res=cv2.resize(img,None,fx=2,fy=2,interpolation=cv2.INTER_CUBIC)
plt.imshow(res)
```

```python
img=cv2.imread("Images/rabbit.png",1)
can=cv2.Canny(img,100,200)
plt.imshow(can)
```

```python
#unsharp masking
from skimage import img_as_float
from skimage.filters import unsharp_mask
img=img_as_float(io.imread("Images/rabbit.png",1))
gaus=gaussian(img,sigma=1,mode="constant",cval=0.0)


new=(gaus-img)*2
final=img+new

fig=plt.figure(figsize=(12,12))
ax1=fig.add_subplot(1,3,1)
ax1.imshow(img,cmap="gray")


ax2=fig.add_subplot(1,3,2)
ax2.imshow(final,cmap="gray")



unsharp= unsharp_mask(img,radius=3,amount=2)

ax3=fig.add_subplot(1,3,3)
ax3.imshow(unsharp,cmap="gray")
ax3.title.set_text("unsharp")


```

```python
#task
img=img_as_float(io.imread("Images/ss.png",1))


gauscv=cv2.GaussianBlur(img,(3,3),0,borderType=cv2.BORDER_CONSTANT)


fig=plt.figure(figsize=(12,12))
ax1=fig.add_subplot(1,3,1)
ax1.imshow(gauscv,cmap="gray")
ax1.title.set_text("Gaussian")



# EDGE DETECTION
#Canny
img2=cv2.imread("Images/ss.png",1)


can=cv2.Canny(img2,100,200)

ax2=fig.add_subplot(1,3,2)
ax2.imshow(can,cmap="gray")
ax2.title.set_text("Canny")

#sobel

sob=sobel(img2)

ax3=fig.add_subplot(1,3,3)
ax3.imshow(sob,cmap="gray")
ax3.title.set_text("Sobel")


```
