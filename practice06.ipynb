```python
import os,cv2
from matplotlib import pyplot as plt
from skimage import io,color,img_as_float
from skimage.transform import rescale,resize,downscale_local_mean
from skimage.filters import gaussian,sobel,median
from skimage.restoration import denoise_bilateral
import numpy as np
```

```python
#median blur

from skimage.morphology import disk
img=cv2.imread("Images/gaussian noise.png",0)
medcv= cv2.medianBlur(img,3)

medsk=median(img,disk(3),mode="constant", cval=0.0)

fig=plt.figure(figsize=(12,12))

ax1=fig.add_subplot(1,2,1)
ax1.imshow(medcv,cmap="gray")

ax2=fig.add_subplot(1,2,2)
ax2.imshow(medsk,cmap="gray")
```

```python
#bilateral

cv_bil=cv2.bilateralFilter(img,5,20,100,borderType=cv2.BORDER_CONSTANT)
sk_bil=denoise_bilateral(img, sigma_color=0.05,sigma_spatial=15,channel_axis=None)
```
