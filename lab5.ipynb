```python
from matplotlib import pyplot as plt
from skimage import io,color,img_as_float
from skimage.transform import rescale,resize,downscale_local_mean
from skimage.filters import gaussian,sobel,unsharp_mask
import cv2
import numpy as np
import os

```

```python
img=io.imread("Images/tiger.jpg", as_gray=True)
```

```python
img_rescaled= rescale(img,1.0/4.0, anti_aliasing=False)
```

```python
plt.imshow(img_rescaled)
```

```python
img_resized=resize(img,(200,100),anti_aliasing=True)
plt.imshow(img_resized)
```

```python
downscale=downscale_local_mean(img,(2,3))
plt.imshow(downscale)
```

```python
gaus=gaussian(img,sigma=1,mode="constant",cval=0.0)
plt.imshow(gaus)
```

```python
sobel=sobel(img)
plt.imshow(sobel)
```

```python
image=cv2.imread("Images/cat.jpeg",1)
```

```python
image.shape
plt.imshow(image)
```

```python
height,width=image.shape[:2]
```

```python
#res = cv2.resize(img,None,fx=2, fy=2,interpolation = cv2.INTER_CUBIC)
res=cv2.resize(image,(width*2,height*2),interpolation=cv2.INTER_LINEAR)
plt.imshow(res)
```

```python
img=cv2.imread("Images/tiger.jpg",1)
image=cv2.imread("Images/cat.jpeg",1)
image=cv2.resize(image,(img.shape[1],img.shape[0]))
```

```python
merged=cv2.addWeighted(img,0.3,image,0.5,0)
plt.imshow(merged)
cv2.imshow("merge",merged)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img=cv2.imread("Images/tiger.jpg",0)
image=cv2.imread("Images/cat.jpeg",0) #needs to be gray
```

```python
edge=cv2.Canny(img,100,200)
cv2.imshow("original",img)
cv2.imshow("edged",edge)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#unsharp mask
#enhanced picture= original+amount*(original-blurred)
```

```python
img1= img_as_float(io.imread("Images/cat.jpeg", as_gray=True))
img2=gaussian(img1,sigma=2,mode="constant",cval=0.0)
img3=2*(img1-img2)
img4=img1+img3
plt.imshow(img4,cmap="gray")
```

```python
unsharp=unsharp_mask(img,radius=3,amount=2)
fig=plt.figure(figsize=(12,12))
ax1=fig.add_subplot(1,2,1)
ax1.imshow(img,cmap="gray")
ax2=fig.add_subplot(1,2,2)
ax2.imshow(unsharp,cmap="gray")
```

```python
img= img_as_float(io.imread("Images/cat.jpeg", as_gray=True))
image= img_as_float(io.imread("Images/tiger.jpg", as_gray=True))
```

```python
gaus_cv2=cv2.GaussianBlur(img,(3,3),0)
gaus_sk=gaussian(img,sigma=1,mode="constant",cval=0.0)
fig=plt.figure(figsize=(12,12))
ax1=fig.add_subplot(1,3,1)
ax1.imshow(img)
ax2=fig.add_subplot(1,3,2)
ax2.imshow(gaus_cv2)
ax3=fig.add_subplot(1,3,3)
ax3.imshow(gaus_sk)

```
